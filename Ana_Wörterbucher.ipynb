```python
import json
from collections import Counter
import re
#Laden der JSON-Datei
with open('jobs_gr_1000_v6.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Extraktion aller Texte aus den Textfeldern
alle_texte = [element['title'] for element in data]

# Zerlegen der Texte in Wörter
woerter = []
for text in alle_texte:
    woerter.extend(re.findall(r'\b\w+\b', text.lower()))

# Gesamtanzahl der Wörter
gesamtanzahl_woerter = len(woerter)

# Zählen der Häufigkeit jedes Wortes
woerter_haeufigkeit = Counter(woerter)

# Ermitteln der 1000 am häufigsten verwendeten Wörter
haufigsten_woerter = woerter_haeufigkeit.most_common(20000)

# Gesamtanzahl der in den 1000 häufigsten Wörtern enthaltenen Wörter
anzahl_woerter_in_top_1000 = sum(anzahl for wort, anzahl in haufigsten_woerter)

# Ausgabe der Gesamtanzahl der Wörter und der Anzahl in den Top 1000
print(f'Gesamtanzahl der Wörter: {gesamtanzahl_woerter}')
print(f'Anzahl der Wörter in den Top 1000: {anzahl_woerter_in_top_1000}')

# Optional: Ausgabe der 1000 häufigsten Wörter
for wort, anzahl in haufigsten_woerter:
     print(f'{wort}')
```

```python
import json

dateipfad = 'woerterbuch_stepstone_w.json'

def entferne_duplikate_und_gebe_geloeschte_zurueck(dateipfad):
    try:
        with open(dateipfad, 'r', encoding='utf-8') as file:
            daten = json.load(file)

        # Entferne Duplikate und speichere gelöschte Einträge
        eintraege_set = set()
        geloeschte_duplikate = []
        for eintrag in daten:
            if eintrag not in eintraege_set:
                eintraege_set.add(eintrag)
            else:
                geloeschte_duplikate.append(eintrag)

        # Schreibe die bereinigten Daten zurück in die Datei
        with open(dateipfad, 'w', encoding='utf-8') as file:
            json.dump(list(eintraege_set), file, ensure_ascii=False, indent=4)

        return geloeschte_duplikate

    except json.JSONDecodeError:
        print("Fehler beim Parsen der JSON-Datei.")
        return []
    except FileNotFoundError:
        print("Datei nicht gefunden.")
        return []
    except Exception as e:
        print(f"Ein Fehler ist aufgetreten: {e}")
        return []

# Aufruf der Funktion
geloeschte_duplikate = entferne_duplikate_und_gebe_geloeschte_zurueck(dateipfad)
if geloeschte_duplikate:
    print("Folgende Duplikate wurden gelöscht:")
    for duplikat in geloeschte_duplikate:
        print(duplikat)
else:
    print("Keine Duplikate gefunden oder ein Fehler ist aufgetreten.")
```

```python
import json

def loesche_gemeinsame_eintraege(dateipfad1, dateipfad2):
    try:
        with open(dateipfad1, 'r', encoding='utf-8') as file1, open(dateipfad2, 'r', encoding='utf-8') as file2:
            daten1 = json.load(file1)
            daten2 = json.load(file2)

        eintraege_set1 = set(daten1)
        eintraege_set2 = set(daten2)

        gemeinsame_eintraege = eintraege_set1.intersection(eintraege_set2)

        # Entferne gemeinsame Einträge aus den Daten
        bereinigte_daten1 = [eintrag for eintrag in daten1 if eintrag not in gemeinsame_eintraege]
        bereinigte_daten2 = [eintrag for eintrag in daten2 if eintrag not in gemeinsame_eintraege]

        # Schreibe die bereinigten Daten zurück in die Dateien
        with open(dateipfad1, 'w', encoding='utf-8') as file1, open(dateipfad2, 'w', encoding='utf-8') as file2:
            json.dump(bereinigte_daten1, file1, ensure_ascii=False, indent=4)
            json.dump(bereinigte_daten2, file2, ensure_ascii=False, indent=4)

        return gemeinsame_eintraege

    except json.JSONDecodeError:
        print("Fehler beim Parsen der JSON-Dateien.")
        return []
    except FileNotFoundError:
        print("Eine der Dateien wurde nicht gefunden.")
        return []
    except Exception as e:
        print(f"Ein Fehler ist aufgetreten: {e}")
        return []

# Beispielaufruf
dateipfad1 = 'woerterbuch_fuehrmint_m.json'
dateipfad2 = 'woerterbuch_fuehrmint_w.json'
gemeinsame_eintraege = loesche_gemeinsame_eintraege(dateipfad1, dateipfad2)
if gemeinsame_eintraege:
    print("Folgende Einträge wurden in beiden Dateien gefunden und gelöscht:")
    for eintrag in gemeinsame_eintraege:
        print(eintrag)
else:
    print("Keine gemeinsamen Einträge gefunden oder ein Fehler ist aufgetreten.")
```
