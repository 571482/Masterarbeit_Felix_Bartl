```python
import spacy
from spacy.matcher import Matcher

nlp = spacy.load("de_core_news_sm")
matcher = Matcher(nlp.vocab)

# Listen mit Berufsbezeichnungen
job_titles = ['Assistent', 'MTA']  # Ersetzen Sie diese durch Ihre vollständige Liste
berufe_femininum = []  # Ihre Liste hier
berufe_gm_plural = []  # Ihre Liste hier

# Gender-Marker
gender_markers = ["*", "/", ":", "_", "·", "I"]

# Funktion, um Muster hinzuzufügen
def add_patterns(list_of_titles, label):
    for title in list_of_titles:
        pattern = [{"LOWER": {"REGEX": title.lower()}}]
        matcher.add(label, [pattern])

# Muster zu Matcher hinzufügen
add_patterns(job_titles, "JOB_TITLES")
add_patterns(berufe_femininum, "FEMININUM")
add_patterns(berufe_gm_plural, "GM_PLURAL")

# Beispieltext
text = "motivierte:n Technische:n Assisstent:in/ MTA für die Erforschung..."
doc = nlp(text)

# Matches finden
matches = matcher(doc)

# Funktion, um Gender-Marker zu finden
def find_gender_markers_nearby(matches, doc, gender_markers):
    for match_id, start, end in matches:
        span = doc[start:end]  # Der gefundene Beruf
        window = doc[max(span.start - 1, 0) : min(span.end + 1, len(doc))]
        for token in window:
            for marker in gender_markers:
                if marker in token.text:
                    print(f"Gender-Marker '{marker}' gefunden in '{span.text}'")

# Gender-Marker in der Nähe von Berufsbezeichnungen suchen
find_gender_markers_nearby(matches, doc, gender_markers)
```
